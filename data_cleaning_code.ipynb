ANOVA Test

```python
from scipy import stats

# Create a list of yield arrays for each river part category
river_part_yields = []
for category in df_t['RIVER_PART'].unique():
    # Filter out zero yields if we want to analyze only productive plots
    # For this initial test, we'll include all yields.
    river_part_yields.append(df_t[df_t['RIVER_PART'] == category]['YIELD'].values)

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*river_part_yields)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("\nConclusion: There is a statistically significant difference in mean yield across different River Part categories.")
else:
    print("\nConclusion: There is no statistically significant difference in mean yield across different River Part categories.")
```



*   Total days in each standard season



```python
# Function for calculating total number of days during standard season
from datetime import datetime
from dateutil.relativedelta import relativedelta

# mapping for month
month_map = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}

def calculate_season_days(season_str, year):

    # split the start month and end month
    start_month_str, end_month_str = season_str.split('-')
    start_month = month_map[start_month_str]
    end_month = month_map[end_month_str]

    # start date of starting month
    start_date = datetime(year, start_month, 1)

    # last date of ending month
    # if end month is earlier than start month we need to consider next year
    if end_month < start_month:
        end_date = datetime(year + 1, end_month, 1) + relativedelta(months=1) - relativedelta(days=1)
    else:
        end_date = datetime(year, end_month, 1) + relativedelta(months=1) - relativedelta(days=1)

    # calucation for total days
    return (end_date - start_date).days + 1
```

```python
# New column for total days in season
df_t['SEASON_DURATION'] = df_t.apply(lambda row: calculate_season_days(row['STANDARD_SEASON'], row['YEAR']), axis=1)
```
